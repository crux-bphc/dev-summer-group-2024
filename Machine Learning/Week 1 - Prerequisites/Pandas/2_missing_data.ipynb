## Missing Data

Most real-world datasets contain _missing data_. This could be beacuse of any number of reasons, some including, inabilty to get all possible data from the whole sample, loss of data, wrong format that had to be deleted etc.

Lets look at how we deal with missing data

```python
import numpy as np
import pandas as pd
```

There are three main methods to deal with this missing data:

1. Drop rows/columns with missing data
2. Fill missing values with something you can infer from the rest of the data
3. A hybrid between the two

```python
# lets import the data
titanic = pd.read_csv('datasets/titanic.csv')
titanic.head()
```

We can already see from just the head that we have NaN values. Missing data is usually filled with NaN. NaN is stored with the help of ``numpy.nan``

### Method 1: Dropping rows/columns with missing values

``dropna()``is a method defined on the ``DataFrame`` class. We specify a ``thresh`` argument the minimum number of non-missing values we want to have in a row/column.

```python
titanic.dropna() # by default it drops all rows with any missing values
```

```python
titanic.dropna(axis=1) # drop columns with missing values
```

Note that these functions dont change the original ``dataFrame``. If you wish to directly modify the original ``dataFrame``, you can put ``inplace = True`` as an argument to the ``dropna()`` function

```python
# titanic.dropna(inplace = True) # drop rows with missing values and update the dataframe
# uncomment and try it out if you want to see how it works
```

### Method 2: Filling the missing values
``fillna()`` is a method defined on the ``DataFrame`` class (and ``Series`` as well). We specify a value argument to fill in the missing values.

```python
titanic.fillna(0) # fill missing values with 0
```

Instead of filling in the missing values with a constant, we can fill in the missing values with a value of some significance, say like the mean or mode of that column.

First lets check which columns have missing values. we can use the ``info()`` method

```python
titanic.info()
```

Since the age column is of numeric datatype, we can find a mean, median or mode for it and fill the NaN values

```python
titanic['Age'].fillna(value=titanic['Age'].mean()) # fill missing values with the mean of the column
```

```python
titanic['Age'].fillna(value=titanic['Age'].median()) # fill missing values with the median of the column
```

```python
titanic['Age'].fillna(value=titanic['Age'].mode()) # fill missing values with the mode of the column
```

**Note**: Other statistics about the dataframe can be found from the describe() method.

```python
titanic.describe()
```

### Example of Method 3: Hybrid (combination of dropna and fillna)

```python
titanic_2 = titanic.dropna(axis=0, thresh=2).copy() # drop rows with at least 2 missing values
titanic_2['Age'].fillna(value=titanic['Age'].mean(), inplace=True) # fill missing values with the mean of the column
```

```python
titanic_2.info()
```

```python
titanic_2.shape
```

**Note**: In the last example we use the ``copy()`` method to create a copy of the dataframe, as by default all the functions return just a view of the dataframe

Summary:
1. ``dropna`` can drop rows/columns decided by the ``axis`` argument
2. ``fillna`` fills the missing values with the argument ``values``. You can fill it with custom values or the mean/median/mode of the column
